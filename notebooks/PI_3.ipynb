```python
import pandas as pd
import numpy as np
import folium

from folium.plugins import HeatMap
from sklearn.cluster import KMeans
```

```python
datos = {
    'lugar': [
        'Centro',
        'Portal del Quindio',
        'Universidad del Quindio',
        'Glorieta Malibu',
        'Hospital San Juan de Dios',
        'Coliseo del Cafe',
        'Centro',
        'Portal del Quindio',
        'Universidad del Quindio',
        'Glorieta Malibu'
    ],

    'latitud': [
        4.5350,
        4.5545,
        4.5551,
        4.5432,
        4.5338,
        4.5510,
        4.5362,
        4.5532,
        4.5561,
        4.5420
    ],

    'longitud': [
        -75.6757,
        -75.6581,
        -75.6615,
        -75.6725,
        -75.6811,
        -75.6670,
        -75.6765,
        -75.6590,
        -75.6622,
        -75.6730
    ],

    'trafico': [
        90,
        70,
        65,
        85,
        60,
        50,
        95,
        75,
        68,
        88
    ]
}
```

```python
df = pd.DataFrame(datos)

# Convertir tipos de datos
df['latitud'] = df['latitud'].astype(float)
df['longitud'] = df['longitud'].astype(float)
df['trafico'] = df['trafico'].astype(int)

df.head()
```

**Valor de tráfico**

| Valor  | Nivel |
| ------ | ----- |
| 0–40   | Bajo  |
| 41–70  | Medio |
| 71–100 | Alto  |


**Crear mapa interactivo**

```python
mapa = folium.Map(
    location=[4.543, -75.673],
    zoom_start=13
)
```

**Agregar puntos al mapa**

```python
mapa = folium.Map(
    location=[4.543, -75.673],
    zoom_start=13
)
for _, fila in df.iterrows():
    folium.CircleMarker(
        location=[
            float(fila['latitud']),
            float(fila['longitud'])
        ],
        radius=float(fila['trafico']) / 10,
        popup=f"Lugar: {fila['lugar']} - Trafico: {int(fila['trafico'])}",
        color='red',
        fill=True,
        fill_color='red'
    ).add_to(mapa)
mapa
```

**Crear Mapa de Calor**

```python
datos_calor = [
    [
        float(fila['latitud']),
        float(fila['longitud']),
        float(fila['trafico'])
    ]
    for _, fila in df.iterrows()
]
mapa_calor = folium.Map(
    location=[4.543, -75.673],
    zoom_start=13
)
HeatMap(datos_calor).add_to(mapa_calor)
mapa_calor
```

**Mapa de Cluster**

```python
from sklearn.cluster import KMeans
```

```python
X = df[['latitud', 'longitud', 'trafico']]
```

```python
kmeans = KMeans(
    n_clusters=3,
    random_state=42
)
df['cluster'] = kmeans.fit_predict(X)
df
```

```python
mapa_cluster = folium.Map(
    location=[4.543, -75.673],
    zoom_start=13
)
colores = {
    0: 'green',
    1: 'orange',
    2: 'red'
}
for _, fila in df.iterrows():
    cluster = int(fila['cluster'])
    folium.CircleMarker(
        location=[
            float(fila['latitud']),
            float(fila['longitud'])
        ],

        radius=10,
        popup=f"""
        Lugar: {fila['lugar']}
        Cluster: {cluster}
        Tráfico: {fila['trafico']}
        """,
        color=colores[cluster],
        fill=True,
        fill_color=colores[cluster],
        fill_opacity=0.7

    ).add_to(mapa_cluster)
mapa_cluster
```

- identifica patrones espaciales automáticamente
- clasifica zonas críticas
- detecta agrupamientos urbanos
- realiza análisis inteligente de movilidad

```python
for cluster in sorted(df['cluster'].unique()):
    promedio = df[df['cluster'] == cluster]['trafico'].mean()
    print(f"\nCLUSTER {cluster}")
    print(f"Promedio de tráfico: {promedio:.2f}")
    if promedio < 60:
        print("Nivel de congestión: BAJO")
    elif promedio < 80:
        print("Nivel de congestión: MEDIO")
    else:
        print("Nivel de congestión: ALTO")
```

Mediante técnicas de clustering espacial utilizando el algoritmo K-Means, el sistema logró identificar agrupaciones territoriales asociadas a distintos niveles de congestión vehicular dentro de la ciudad de Armenia. Esto permitió detectar zonas críticas y sectores con comportamiento vial similar.
